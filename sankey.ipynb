```python
# Helper function
def get_keys_by_value(dictionary, value):
    keys_with_value = []
    for key, val in dictionary.items():
        if val == value:
            keys_with_value.append(key)
    return keys_with_value
```

```python
named_link_source = []
named_link_target = []
link_value = []
organizations = []
```

```python
# Get complete list of organizations for 07 and 09
all_organizations_07 = dict()
for i in range(1, df_07_in_person.shape[1] + 1):
    set_07 = set(df_07_in_person['Group ' + str(i)])
    if '0' in set_07:
        set_07.remove('0')
    for organization in set_07:
        all_organizations_07[organization] = 'Group ' + str(i)

all_organizations_09 = dict()
for i in range(1, df_09_in_person.shape[1] + 1):
    set_09 = set(df_09_in_person['Group ' + str(i)])
    if '0' in set_09:
        set_09.remove('0')
    for organization in set_09:
        all_organizations_09[organization] = 'Group ' + str(i)
```

```python
# [TESTING] Flow between 07 and 09, 07 and 14 (without 09), 07 and 17 (without 09 and 14)
for i in range(1, df_07_in_person.shape[1] + 1):
    
    # Get all the organizations in all_organizations_07 with value Group i
    set_07 = set(get_keys_by_value(all_organizations_07, 'Group ' + str(i)))

    for j in range(1, df_09_in_person.shape[1] + 1):
        set_09 = set(df_09_in_person['Group ' + str(j)])
        if '0' in set_09:
            set_09.remove('0')
        flow = set_07 & set_09
        if len(flow) > 0:
            # remove set_07 from complete list of 07 organizations
            # print(str(i) + ', ' + str(j) + ': ' + str(flow))
            named_link_source.append('07 Group ' + str(i))
            named_link_target.append('09 Group ' + str(j))
            link_value.append(len(flow))
            organizations.append(flow)
            set_07.difference_update(flow)

    for k in range(1, df_14_in_person.shape[1] + 1):
        set_14 = set(df_14_in_person['Group ' + str(k)])
        if '0' in set_14:
            set_14.remove('0')
        flow = set_07 & set_14
        if len(flow) > 0:
            named_link_source.append('07 Group ' + str(i))
            named_link_target.append('14 Group ' + str(k))
            link_value.append(len(flow))
            organizations.append(flow)
            set_07.difference_update(flow)
    
    for l in range(1, df_17_in_person.shape[1] + 1):
        set_17 = set(df_17_in_person['Group ' + str(l)])
        if '0' in set_17:
            set_17.remove('0')
        flow = set_07 & set_17
        if len(flow) > 0:
            named_link_source.append('07 Group ' + str(i))
            named_link_target.append('17 Group ' + str(l))
            link_value.append(len(flow))
            organizations.append(flow)
            set_07.difference_update(flow)
```

```python
# [TESTING] Flow between 09 and 14, 09 and 17 (without 14)
for i in range(1, df_09_in_person.shape[1] + 1):
    
    # Get all the organizations in all_organizations_09 with value Group i
    set_09 = set(get_keys_by_value(all_organizations_09, 'Group ' + str(i)))

    for j in range(1, df_14_in_person.shape[1] + 1):
        set_14 = set(df_14_in_person['Group ' + str(j)])
        if '0' in set_14:
            set_14.remove('0')
        flow = set_09 & set_14
        if len(flow) > 0:
            named_link_source.append('09 Group ' + str(i))
            named_link_target.append('14 Group ' + str(j))
            link_value.append(len(flow))
            organizations.append(flow)
            set_09.difference_update(flow)

    for k in range(1, df_17_in_person.shape[1] + 1):
        set_17 = set(df_17_in_person['Group ' + str(k)])
        if '0' in set_17:
            set_17.remove('0')
        flow = set_09 & set_17
        if len(flow) > 0:
            named_link_source.append('09 Group ' + str(i))
            named_link_target.append('17 Group ' + str(k))
            link_value.append(len(flow))
            organizations.append(flow)
            set_09.difference_update(flow)
```

```python
# Flow between 14 and 17
for i in range(1, df_14_in_person.shape[1] + 1):
    
    set_14 = set(df_14_in_person['Group ' + str(i)])
    if '0' in set_14:
        set_14.remove('0')

    for j in range(1, df_17_in_person.shape[1] + 1):
        set_17 = set(df_17_in_person['Group ' + str(j)])
        if '0' in set_17:
            set_17.remove('0')
        
        flow = len(set_14 & set_17)
        if flow > 0:
            # print(str(i) + ', ' + str(j) + ': ' + str(flow))
            named_link_source.append('14 Group ' + str(i))
            named_link_target.append('17 Group ' + str(j))
            link_value.append(flow)
            organizations.append(set_14 & set_17)
```
